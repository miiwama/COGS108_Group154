```python
import pandas as pd
df1 = pd.read_csv('merged_sdge_cleaned.csv')
df2 = pd.read_csv('comp_weather_data.csv')
df2.rename(columns={'zipcode': 'ZipCode', 'month': 'Month', 'year': 'Year'}, inplace=True)
df2['ZipCode'] = df2['ZipCode'].astype(int)
df1['ZipCode'] = df1['ZipCode'].astype(int)
merged_df = pd.merge(df1, df2, on=['ZipCode', 'Month', 'Year'], how='inner')
merged_df.to_csv('merged_energy_weather_data.csv', index=False)
print(merged_df.head())
```
