## Data overview

This study will analyze variables such as zip codes, monthly average electricity consumption, average temperature, humidity, precipitation, date, month, and year. Data will be aggregated at the zip code level across San Diego County. The analysis will focus on the period from 2021 to the present, excluding data from 2020 due to the pandemic’s impact on energy consumption behaviors. Since people were largely confined to their homes in 2020, electricity usage patterns were atypical and could skew the results. Similarly, we will exclude pre-pandemic data, as energy consumption habits may have been significantly different prior to 2020. Beginning the analysis in 2021 allows us to establish a more stable and representative baseline that reflects post-pandemic adjustments.
